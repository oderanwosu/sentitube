```python

import nltk
from IPython.display import display
from  yt_api import get_video_comments_data
from yt_api import get_video_title_data
import pandas as pd
import matplotlib.pyplot as plt
import timeago, datetime
import numpy as np
```

```python
#enter a youtube video's ID
id1 = ''
id2 = ''
video1 = {'df': pd.DataFrame(get_video_comments_data(id1)), 'meta': get_video_title_data(id1)}
video2 = {'df': pd.DataFrame(get_video_comments_data(id2)), 'meta': get_video_title_data(id2)}
```

```python
video1['df']['color'] = ['r' for i in video1['df']['dn']]
video2['df']['color'] = ['b' for i in video2['df']['dn']]
df = pd.concat([video1['df'], video2['df']])
df.sort_values(by='published_at', inplace=True )
```

```python

plt.scatter(df['published_at'], df['compound'],  df['like_counts'], c=df['color'], alpha=0.5)
plt.xticks(df['published_at'])
# plt.plot(df1['published_at'], df1['compound'], c='b') 
# plt.plot(df2['published_at'], df2['compound'], c='r')

plt.title('Youtube Video Comments over time Sentiment Analysis Scatterplot ')
plt.ylabel('Sentiment score')
plt.xlabel('Time')
plt.show()


```
